```python
class LemmaTokenizer:
    ignore_tokens = [',', '.', ';', ':', '"', '``', "''", '`']
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
      res=[]
      for token in word_tokenize(doc):
        word=''.join(filter(str.isalnum, token))
        if ((word not in self.ignore_tokens) and (len(self.wnl.lemmatize(word))>2) and not(any(char.isdigit() for char in word))):
          res.append(self.wnl.lemmatize(word))
      return res

stop_words = nltk.corpus.stopwords.words('english')
#stop_words.extend(["more","words"])

tokenizer = LemmaTokenizer()
token_stop = list(set(tokenizer(' '.join(stop_words))))
token_stop.sort()
```

```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import nltk

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('english')
#stop_words.extend(["more","words"])
stop_words.sort()
```

```python
subj_report_5 = subj_report[:5]
```

```python
punctuation = [',', '.', ';', ':', '"', "-", "@", "#"
               '`', '´', "’", "'", "!", "?", "(", ")",
               "/", "$", "%"]


concat = []

wnl = WordNetLemmatizer()

rejected_words = {}
replacement_dict = {}
rejected_lists = {}

for txt in subj_report:

  list_words = [wnl.lemmatize(i.lower(),j[0].lower()) if j[0].lower() in ['a','n','v'] else wnl.lemmatize(i.lower()) for i,j in pos_tag(word_tokenize(txt))]
  
  remaining_words  = []
  for word in list_words:
    
    # might not need to use the two lines below since wnl already did that?
    # for punct in punctuation:
    #   word.replace(punct,"")

    word = word.lower()
    reason = ""
    if any(char.isdigit() for char in word): reason = "Number"
    elif d.check(word)==False: reason = "Not a real word"
    elif word in stop_words: reason = "Stop word"
    elif len(word)<3: reason = "Too short"

    else:
      in_word = word
      out_word = stemmer.stem(in_word)
      
      remaining_words.append(out_word)
      replacement_dict.setdefault(out_word, set([])).update([in_word])

    if reason !="":
      rejected_words.setdefault(word, set([reason]))
      rejected_lists.setdefault(reason, set([])).update([word])

  concat.append(remaining_words)
```

TruncatedSVD

```python
from sklearn.decomposition import TruncatedSVD
```

```python
#number_dimensions = 4
number_dimensions = 194 #coba cari yang optimal, try using silhouette coefficient (ratio between inter cluster indexed and outer cluster indexed)
#hasil hyperparameter tuning, n_components berapapun gak affect performance
lsa_model = TruncatedSVD(n_components = number_dimensions)
lsa_comps = lsa_model.fit_transform(tfidf_array)
print(lsa_comps)
```

```python
type(lsa_comps)
```

```python
print(sum(lsa_model.explained_variance_ratio_))
print(sorted(lsa_model.explained_variance_ratio_))
#makin lama makin kecil (kalo divisualize, linenya naik terus turun terus)
```

Elbow method

```python
explained_variance = lsa_model.explained_variance_ratio_
```

```python
plt.plot(range(1, len(explained_variance)+1), explained_variance, 'bo-', linewidth=2)
plt.xlabel('Number of components')
plt.ylabel('Explained variance ratio')
plt.title('Scree Plot')
plt.show()
```

```python
# Calculate the second derivative of the explained variance ratio curve
second_der = np.diff(explained_variance, 2)

# Find the index of the maximum value of the second derivative
elbow_index = np.argmax(second_der) + 1

# The optimal number of components is the index of the elbow point
n_components_optimal = elbow_index
```

```python
n_components_optimal
```

```python
X_reduced = lsa_model.transform(tfidf_array)[:, :n_components_optimal]
```
