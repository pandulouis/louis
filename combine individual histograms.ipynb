```python
from PIL import Image
```

```python
histogram_files = ['error analysis knn tfidf thc.png', 'error analysis knn tfidf cbd.png', 'error analysis knn tfidf cbn.png',
                   'error analysis knn tfidf cbg.png', 'error analysis knn tfidf cbc.png', 'error analysis knn tfidf thcv.png',
                   'error analysis knn tfidf dlim.png', 'error analysis knn tfidf bocim.png', 'error analysis knn tfidf bmyrc.png',
                   'error analysis knn tfidf bpine.png', 'error analysis knn tfidf linalol.png', 'error analysis knn tfidf alpine.png',
                   'error analysis knn tfidf camphene.png', 'error analysis knn tfidf 3car.png', 'error analysis knn tfidf alterp.png',
                   'error analysis knn tfidf ocim.png', 'error analysis knn tfidf euca.png', 'error analysis knn tfidf terpi.png',
                   'error analysis knn tfidf isopul.png', 'error analysis knn tfidf geraniol.png', 'error analysis knn tfidf guai.png',
                   'error analysis knn tfidf carox.png', 'error analysis knn tfidf albis.png']  # Add all the filenames here


num_rows = 4
num_columns = 6

image_width, image_height = Image.open(histogram_files[0]).size

combined_width = num_columns * image_width
combined_height = num_rows * image_height

combined_image = Image.new('RGB', (combined_width, combined_height))

for i, file in enumerate(histogram_files):
    image = Image.open(file)

    row = i // num_columns
    col = i % num_columns
    x = col * image_width
    y = row * image_height

    combined_image.paste(image, (x, y))

combined_image.save('error combined knn tfidf.png')
```
