```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

model = model.to('cuda')

def bert_encode(text):
    input_ids = torch.tensor(tokenizer.encode(text, add_special_tokens=True))
    padded_input_ids = torch.cat([input_ids[:512], torch.zeros(max(512 - len(input_ids), 0), dtype=torch.long)])
    padded_input_ids = padded_input_ids.to('cuda')
    
    with torch.no_grad():
        outputs = model(torch.unsqueeze(padded_input_ids, dim=0))
        last_hidden_states = outputs[0]
    last_hidden_states = last_hidden_states.to('cpu')
    return torch.mean(last_hidden_states, dim=1).squeeze()

df_bert_final['bert_embeddings'] = df_bert_final['report and species'].apply(bert_encode)

embeddings_array = np.stack(df_bert_final['bert_embeddings'].values)
new_df = pd.DataFrame(embeddings_array, columns=[f'feature_{i}' for i in range(768)])
df_bert_final = pd.concat([df_bert_final, new_df], axis=1)

print(df_bert_final)

```
